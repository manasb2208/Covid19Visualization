**Importing Library function**

```
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```
#Dataframe Aquiring:
df = pd.read_csv("/content/covid19dec3.csv")
df
```

The above seen dataframe is of covid19 situation of Indian state, recorded and updated on 3rd December,2020 on MohFW website.

```
df1 = pd.read_csv("/content/covid19dec13.csv")
df1
```

```
print(df.info())
print("**********************************************")
print(df1.info())
```

There is no need of seeing daily changes on our visualization so we will simply drop those columns which are showing daily changes in active,cured and death cases of states. 

```
df = df.drop(columns= ['AC_Change','Cured_Change','Death_Change'], axis=1)
df
```

```
df1 = df1.drop(columns = ['AC_Change','Cured_Change','Death_Change'], axis = 1)
df1
```

Now as states' names are long enough so we will define e function to trim the names.

```
def trim(x):
  x = x[:5]
  return x
```

```
df['Name_of_State/UT'] = df['Name_of_State/UT'].apply(trim)
df1['Name_of_State/UT'] = df1['Name_of_State/UT'].apply(trim)
```

```
#Highest Active case on 3rd December
print("Highest Active case on 3rd December")
print(df['Active Cases'].max())
print(df[df['Active Cases'] == df['Active Cases'].max()].drop(columns= ['Cured','Deaths']))

print("*********************************************")
#Highest Active case on 13th December
print("Highest Active case on 13th December")
print(df1['Active Cases'].max())
print(df1[df1['Active Cases'] == df1['Active Cases'].max()].drop(columns= ['Cured','Deaths']))
```

```
#Lowest Active case on 3rd December
print("Lowest Active case on 3rd December")
print(df['Active Cases'].min())
print(df[df['Active Cases'] == df['Active Cases'].min()].drop(columns= ['Cured','Deaths']))

print("*********************************************")
print("Lowest Active case on 13th December")
print(df1['Active Cases'].min())
print(df[df1['Active Cases'] == df1['Active Cases'].min()].drop(columns= ['Cured','Deaths']))
```

So now we will represent the state wise active cases through barplot for both the dataset.

```
fig,axes = plt.subplots(2,1, figsize =(90,40))
plt.rcParams.update({'font.size': 40})
axes[0].set_title('Barplot of Statewise active cases')
sns.barplot('Name_of_State/UT', 'Active Cases', data = df, ax = axes[0])

plt.rcParams.update({'font.size': 40})
axes[1].set_title('Barplot of Statewise active cases')
sns.barplot('Name_of_State/UT', 'Active Cases', data = df1, ax = axes[1])
```

From the above two barplot of covid data of two different day in a gap of 10 days we can see highest active case drops down from 89000 to 74000 for Maharastra.

Now we are adding three more column in both of our dataset. They are total infected count and then other two are statewise cured rate and death rate in percentage. 

```
df['Total Infected'] = df['Active Cases'] + df['Cured'] + df['Deaths']
df1['Total Infected'] = df1['Active Cases'] + df1['Cured'] + df1['Deaths']
```

```
df['Cured Rate in percent'] = (df['Cured']*100)/df['Total Infected']
df1['Cured Rate in percent'] = (df1['Cured']*100)/df1['Total Infected']
```

```
df['Death rate in percent'] =  (df['Deaths']*100)/df['Total Infected']
df1['Death rate in percent'] =  (df1['Deaths']*100)/df1['Total Infected']
df
```
